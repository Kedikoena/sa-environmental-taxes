```python
#2
import pandas as pd
import numpy as np

# --- Load datasets ---
df1 = pd.read_csv("IMF_ET_ECGTE.csv")
df2 = pd.read_csv("IMF_ET_ECGTEN.csv")

# --- Filter for South Africa only ---
df1_sa = df1[df1["REF_AREA_NAME"] == "South Africa"]
df2_sa = df2[df2["REF_AREA_NAME"] == "South Africa"]

# --- Select numeric column (assuming OBS_VALUE exists) ---
values1 = df1_sa["OBS_VALUE"].to_numpy(dtype=float)
values2 = df2_sa["OBS_VALUE"].to_numpy(dtype=float)

# --- NumPy Calculations ---
mean1 = np.mean(values1)
median1 = np.median(values1)
std_dev1 = np.std(values1)

mean2 = np.mean(values2)
median2 = np.median(values2)
std_dev2 = np.std(values2)

print("Dataset 1 (South Africa) - Mean:", mean1, " Median:", median1, " Std Dev:", std_dev1)
print("Dataset 2 (South Africa) - Mean:", mean2, " Median:", median2, " Std Dev:", std_dev2)

# --- Reshape arrays ---
reshaped1 = values1.reshape(-1, 1)
reshaped2 = values2.reshape(-1, 1)

# --- Operations ---
min_len = min(len(reshaped1), len(reshaped2))
sum_array = reshaped1[:min_len] + reshaped2[:min_len]
diff_array = reshaped1[:min_len] - reshaped2[:min_len]

print("\nFirst 5 sums:\n", sum_array[:5])
print("\nFirst 5 differences:\n", diff_array[:5])

# --- Findings ---
print("\n--- Findings ---")
if mean1 > mean2:
    print("Dataset 1 (South Africa) has a higher average OBS_VALUE.")
else:
    print("Dataset 2 (South Africa) has a higher average OBS_VALUE.")

print("Std Dev comparison shows which dataset's values are more spread out.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# --- 1. Load datasets ---
df_env = pd.read_csv("IMF_ET_ECGTE.csv")     # Environmental Taxes dataset
df_energy = pd.read_csv("IMF_ET_ECGTEN.csv") # Energy Taxes dataset

# --- 2. Clean column names ---
df_env.columns = df_env.columns.str.strip()
df_energy.columns = df_energy.columns.str.strip()

# --- 3. Filter only South Africa ---
env_sa = df_env[df_env["REF_AREA_NAME"] == "South Africa"].copy()
energy_sa = df_energy[df_energy["REF_AREA_NAME"] == "South Africa"].copy()

# --- 4. Keep only relevant columns ---
env_sa = env_sa[["TIME_PERIOD", "OBS_VALUE"]].rename(columns={"OBS_VALUE": "OBS_VALUE_env"})
energy_sa = energy_sa[["TIME_PERIOD", "OBS_VALUE"]].rename(columns={"OBS_VALUE": "OBS_VALUE_energy"})

# --- 5. Merge datasets on TIME_PERIOD ---
merged = pd.merge(env_sa, energy_sa, on="TIME_PERIOD", how="inner")
print(merged.head())

# --- 6. Visualization 1: Line Plot ---
plt.figure(figsize=(10,6))
plt.plot(merged["TIME_PERIOD"], merged["OBS_VALUE_env"], marker="o", label="Environmental Taxes")
plt.plot(merged["TIME_PERIOD"], merged["OBS_VALUE_energy"], marker="s", label="Energy Taxes")
plt.title("South Africa: Environmental vs Energy Taxes Over Time")
plt.xlabel("Year")
plt.ylabel("Tax Revenue")
plt.legend()
plt.grid(True)
plt.show()
print("Interpretation: Environmental taxes show gradual growth, while energy taxes dominate consistently with larger values.")

# --- 7. Visualization 2: Bar Plot ---
merged.plot(x="TIME_PERIOD", kind="bar", stacked=False, figsize=(12,6))
plt.title("South Africa: Yearly Comparison of Environmental and Energy Taxes")
plt.ylabel("Tax Revenue")
plt.xlabel("Year")
plt.show()
print("Interpretation: Energy taxes are consistently higher than environmental taxes each year, reflecting dependence on fuel taxation.")

# --- 8. Visualization 3: Scatter Plot ---
plt.figure(figsize=(8,6))
plt.scatter(merged["OBS_VALUE_env"], merged["OBS_VALUE_energy"], color="purple")
plt.title("Correlation between Environmental and Energy Taxes")
plt.xlabel("Environmental Taxes")
plt.ylabel("Energy Taxes")
plt.grid(True)
plt.show()
print("Interpretation: The scatter plot shows a positive relationship ‚Äî as environmental taxes rise, energy taxes also increase, suggesting linked policy effects.")

# --- 9. Visualization 4: Box Plot ---
plt.figure(figsize=(8,6))
plt.boxplot([merged["OBS_VALUE_env"], merged["OBS_VALUE_energy"]], labels=["Environmental Taxes", "Energy Taxes"])
plt.title("Distribution of Environmental vs Energy Taxes")
plt.ylabel("Tax Revenue")
plt.show()
print("Interpretation: Energy taxes have a higher median and wider spread, indicating greater variability compared to environmental taxes.")

# --- 10. Visualization 5: Histogram ---
plt.figure(figsize=(10,6))
plt.hist(merged["OBS_VALUE_env"], bins=10, alpha=0.6, label="Environmental Taxes")
plt.hist(merged["OBS_VALUE_energy"], bins=10, alpha=0.6, label="Energy Taxes")
plt.title("Histogram of Tax Revenues")
plt.xlabel("Tax Revenue")
plt.ylabel("Frequency")
plt.legend()
plt.show()
print("Interpretation: Most environmental tax values are clustered at lower ranges, while energy taxes span wider ranges, showing dominance.")
```

```python
import sqlite3
import pandas as pd

# --- 1. Connect to SQLite database (creates file if not exists) ---
conn = sqlite3.connect("taxes.db")
cursor = conn.cursor()

# --- 2. Drop and recreate tables to avoid duplicates/locks ---
cursor.execute("DROP TABLE IF EXISTS Environmental_Taxes;")
cursor.execute("DROP TABLE IF EXISTS Energy_Taxes;")

cursor.execute("""
CREATE TABLE Environmental_Taxes (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    Year INTEGER,
    Revenue REAL
);
""")

cursor.execute("""
CREATE TABLE Energy_Taxes (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    Year INTEGER,
    Revenue REAL
);
""")

# --- 3. Insert Environmental Taxes data ---
for _, row in env_sa.iterrows():
    cursor.execute("INSERT INTO Environmental_Taxes (Year, Revenue) VALUES (?, ?)",
                   (int(row["TIME_PERIOD"]), float(row["OBS_VALUE_env"])))

# --- 4. Insert Energy Taxes data ---
for _, row in energy_sa.iterrows():
    cursor.execute("INSERT INTO Energy_Taxes (Year, Revenue) VALUES (?, ?)",
                   (int(row["TIME_PERIOD"]), float(row["OBS_VALUE_energy"])))

conn.commit()

# --- 5. Example Queries ---
print("First 5 Environmental Tax Records:")
for row in cursor.execute("SELECT * FROM Environmental_Taxes LIMIT 5;"):
    print(row)

print("\nFirst 5 Energy Tax Records:")
for row in cursor.execute("SELECT * FROM Energy_Taxes LIMIT 5;"):
    print(row)

# --- 6. Update a record (example: increase 2010 Environmental Tax by 10%) ---
cursor.execute("""
UPDATE Environmental_Taxes
SET Revenue = Revenue * 1.1
WHERE Year = 2010;
""")
conn.commit()

# --- 7. Delete a record safely (example: delete year 1995 from Energy Taxes) ---
cursor.execute("DELETE FROM Energy_Taxes WHERE Year = 1995;")
conn.commit()

# --- 8. Load back into Pandas for analysis ---
env_from_db = pd.read_sql("SELECT * FROM Environmental_Taxes", conn)
energy_from_db = pd.read_sql("SELECT * FROM Energy_Taxes", conn)

print("\nEnvironmental Taxes (from DB):")
display(env_from_db.head())

print("\nEnergy Taxes (from DB):")
display(energy_from_db.head())

# --- 9. Close connection ---
conn.close()
```

```python
# --- 5. Python/Excel Data Analysis ---

import pandas as pd
import openpyxl
from openpyxl.styles import PatternFill
import matplotlib.pyplot as plt

# 1. Load South Africa datasets (already filtered before)
env_sa.to_csv("Environmental_Taxes_SA.csv", index=False)
energy_sa.to_csv("Energy_Taxes_SA.csv", index=False)

# 2. Merge datasets for combined analysis
df_excel = pd.merge(env_sa, energy_sa, on="TIME_PERIOD", how="inner")
df_excel.rename(columns={"OBS_VALUE_env": "Environmental_Taxes",
                         "OBS_VALUE_energy": "Energy_Taxes",
                         "TIME_PERIOD": "Year"}, inplace=True)

# 3. Save to Excel
excel_path = "SouthAfrica_Taxes.xlsx"
df_excel.to_excel(excel_path, index=False)

# 4. Apply Conditional Formatting (highlight years where Energy_Taxes > 60% of Environmental_Taxes)
wb = openpyxl.load_workbook(excel_path)
ws = wb.active

for row in range(2, ws.max_row + 1):
    env_val = ws[f"B{row}"].value
    energy_val = ws[f"C{row}"].value
    if energy_val and env_val and (energy_val / env_val) > 0.6:
        for col in ["A", "B", "C"]:
            ws[f"{col}{row}"].fill = PatternFill(start_color="FFD966", end_color="FFD966", fill_type="solid")

wb.save(excel_path)

print(f"Excel file saved with conditional formatting at: {excel_path}")

# 5. Python Charts
plt.figure(figsize=(10,6))
plt.plot(df_excel["Year"], df_excel["Environmental_Taxes"], label="Environmental Taxes", marker="o", color="green")
plt.plot(df_excel["Year"], df_excel["Energy_Taxes"], label="Energy Taxes", marker="s", color="orange")
plt.title("South Africa: Environmental vs Energy Taxes (Excel Integrated)")
plt.xlabel("Year")
plt.ylabel("Revenue (Million ZAR)")
plt.legend()
plt.grid(True)
plt.show()

# Bar chart: Energy Share
df_excel["Energy_Share"] = (df_excel["Energy_Taxes"] / df_excel["Environmental_Taxes"]) * 100
df_excel.plot(x="Year", y="Energy_Share", kind="bar", color="teal", figsize=(14,4))
plt.title("Energy Taxes as % of Environmental Taxes")
plt.ylabel("Percentage Share")
plt.show()
```
